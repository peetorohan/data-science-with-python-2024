```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = (r"C:\Users\ivanf\OneDrive\Desktop\Data\P Data.zip")  # Replace with your actual file path
df = pd.read_csv(file_path)

# Display the first few rows
display(df.head())
```

```python
print("Dataset Info:")
df.info()
print("\nSummary Statistics:")
print(df.describe())
```

```python
df.dropna(inplace=True)  # Remove missing values (modify as needed)
```

```python
top_counties = df.nlargest(10, "Total Population")
plt.figure(figsize=(12, 6))
sns.barplot(x=top_counties["Total Population"], y=top_counties["County"], palette="coolwarm")
plt.xlabel("Total Population")
plt.ylabel("County")
plt.title("Top 10 Counties by Population")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Male Population"], bins=20, kde=True, label="Male", color="blue", alpha=0.6)
sns.histplot(df["Female Population"], bins=20, kde=True, label="Female", color="red", alpha=0.6)
plt.xlabel("Population")
plt.ylabel("Counties")
plt.title("Gender Distribution Across Counties")
plt.legend()
plt.show()
```
